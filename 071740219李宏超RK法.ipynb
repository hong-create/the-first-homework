```python
import math                                        
import numpy as np                                 
import matplotlib.pyplot as plt                    
```

```python
g = 9.8

init_z = State(y=100.0 , v=0.0 )
init_x = State(y=0.0 , v=500.0 )

t_0=0
t_end = 6 
dt = 1 

system_z = System(init=init_z, t_0=t_0,g=g, t_end=t_end, dt=dt)
system_x = System(init=init_x,t_0=t_0, t_end=t_end, dt=dt)

def slope_func_z(state, t, system):
    y, v = state
    
    dydt = v
    dvdt = -system.g
    
    return dydt, dvdt

def slope_func_x(state, t, system):
    y, v = state
    
    dydt = v
    dvdt = 0.0
    
    return dydt, dvdt

Vz, Az = slope_func_z(system_z.init, 0, system_z)
Vx, Ax = slope_func_x(system_x.init, 0, system_x)
```

```python
def runge_kutta(y, x, h, f):                      
    """ y 函数初值
        x 自变量初值
        h 自变量步长
        f 函数求导
    """                                            
    k1 = h * f(y, t)
    k2 = h * f(y + 0.5 * k1, x + 0.5 * h)
    k3 = h * f(y + 0.5 * k2, x + 0.5 * h)
    k4 = h * f(y + k3, x + h)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6.    
```

```python
#考察竖直方向的自由落体运动
if __name__=='__main__':                           
    t = 0.                                                                             
    y = 100.
    dt = 0.01
    ys, ts = [], []                                
    def func(y, t):                                
        return  -1*math.sqrt(2*g*y)
    while t <= 6:
        y = runge_kutta(y, t, dt, func)            
        t += dt
        ys.append(y)
        ts.append(t)                               
```

```python
exact = [100-0.5*g*t*t for t in ts]      #函数举例
```

```python
plt.plot(ts, ys, label='runge_kutta')          #分别画出函数图像
plt.legend()
plt.show()
plt.plot(ts, exact, label='exact')
plt.legend()
plt.show()
plt.plot(ts, ys, label='runge_kutta')          #在一个坐标系下画出函数图像
plt.plot(ts, exact, label='exact')
plt.legend()
plt.show()
```
